## Clean the data

```python
# find rows with missing values
na_rows = df[df.iloc[:, 4:6].isin(['na', 'N.A.']).any(axis=1)].index # na values are stored as 'na' or 'N.A.' strings

# drop rows
df.drop(na_rows, inplace=True)
df2 = df2.dropna()

# Change univeristy names to their abbreviations
def acronym(text):
    return ''.join(word[0] for word in text.split() if word[0].isupper())

df['university'] = df['university'].apply(acronym)

# combine new and old data
df = pd.concat([df, df2])

# convert datatype of numeric columns
cols_float = df.columns[4:6]
cols_int   = df.columns[6:12]
df[cols_float] = df[cols_float].astype('float')
df[cols_int] = df[cols_int].astype('float').astype('int')

# export cleaned dataset
df.to_csv(r'C:\Users\owner\Downloads\wrangledGES.csv', index=False) # export dataframe as csv for visualisation later on
```
