## Drop rows with missing values

```python
na_rows = df[df.iloc[:, 4:6].isin(['na', 'N.A.']).any(axis=1)].index # na values are stored as 'na' or 'N.A.' strings
df.drop(na_rows, inplace=True)
df.shape[0] # check whether rows are dropped
```

## Change datatype of numeric columns

```python
cols_float = df.columns[4:6]
cols_int   = df.columns[6:12]
df[cols_float] = df[cols_float].astype('float')
df[cols_int] = df[cols_int].astype('float').astype('int')
```

```python
df.info() # check whether data types are changed
```

## Convert uni names to their acronyms

```python
def acronym(text):
    return ''.join(word[0].upper() for word in text.split())

df['university'] = df['university'].apply(acronym)
```
